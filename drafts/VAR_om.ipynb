# VECTOR AUTOREGRESSION with an offset

This notebook aims to esplore some axpects of *VAR* in order to understand why the notebooks on SLDS inference do not work.
The inference will be written in homogeneous coordinates.

```python
import numpy as np
import matplotlib.pyplot as plt
from sklearn.datasets import make_spd_matrix  #to generate covariance matrices
from scipy.stats import special_ortho_group
from numpy.random import dirichlet
from scipy.stats import invwishart, matrix_normal
from numpy.linalg import inv,det
from numpy import sqrt
from tqdm import trange

seed = 123
np.random.seed(seed)
```

## Data Generation
Let's start generating data using an equation of the kind $x_t = Ax_{t-1} + b + \xi$, $\xi \sim \operatorname{N(0,Q)}$, $b$ time independent.

```python
# number of observable variables (coordinates of a 2-D motion)
M = 2

#number of samples you want to generate
N_samples = 1000

A_gen =  special_ortho_group.rvs(M)
Q_gen =  make_spd_matrix(M, random_state = seed)
b_gen = np.random.randn(M)
```

```python
#### initializing the x time-serie
x_serie = np.zeros((M,N_samples))


for n in np.arange(N_samples - 1):
    
    #drawing noise from a normal
    noise = np.random.multivariate_normal(mean = np.zeros(M), cov = Q_gen, size = 1)
   
    x_serie[:,n + 1] = A_gen @ x_serie[:,n] + b_gen + noise.flatten()
    
fig,ax = plt.subplots(1,1,figsize = [15,15])

ax.plot(x_serie[0,:], x_serie[1,:], marker = '.', color = 'black') 
ax.scatter(x_serie[0,-1], x_serie[1,-1], color = 'red', s = 40)
ax.scatter(x_serie[0,0], x_serie[1,0], color = 'blue', s= 40)
```

```python
print('A matrix:','\n', A_gen)
print('Q matrix:','\n', Q_gen)
print('b vector:','\n', b_gen)
```

## DATA PREPARATION

```python
T = N_samples

#preparing data
X = x_serie[:, 0:T-1]          #this X_{t-1} so times run between [0,T-1]

Y = x_serie[:, 1:T]            #this is X_t so times run between [1,T], dimensions M x (T-1)

n_iterations = 2000            #number of iterations in the Gibbs sampler
burn_out = 500

#voglio seguire in tutto e per tutto le formule su wikipedia
Y = Y.T
X = X.T

print('X mean:',  X.mean(axis = 0),'\n')
print('Y mean:',  Y.mean(axis = 0),'\n')

'''
#maybe here it is necessary to subtract the mean because I am considering no offset
X = X - X.mean(axis = 0)
Y = Y - Y.mean(axis = 0)
'''

#adding a column of ones to X.T
X = np.column_stack((X,np.ones(T-1)))                    #dimensions (T-1)x(M+1)

#ora A è (M+1)xM e la sua ultima riga è b, Y è (T-1)xM, X è (T-1)x(M+1) 
print('X dimensions:', X.shape,'\n')
print('Y dimensions:', Y.shape,'\n')
```

## INFERENCE

```python
#initialising

Q = np.zeros((n_iterations, M, M))            #covariance matrix
L = np.zeros((n_iterations, M+1, M+1))        #lambda parameter for matrix normal
V = np.zeros((n_iterations, M, M))            #V parameter for the inverse Wishart
nu = np.ones(n_iterations)*M                  #nu parameter for inverse wishart, intialization wrong in inference_2
A = np.zeros((n_iterations, M+1, M)) 
Mn = np.zeros((n_iterations, M+1, M))         #mean matrix for matrix normal
             
#priors
             
L[0] = make_spd_matrix(M+1, random_state = seed)                      #I am not sure it is necessary to keep all the iterations
V[0] = make_spd_matrix(M, random_state = seed)
                                                                    #Mn has already been initialised
Q[0] = invwishart.rvs(df = nu[0], scale = V[0])                     #first approximation of variance matrix
A[0] = matrix_normal.rvs(mean = Mn[0],\
                                   rowcov = inv(L[0]), colcov = Q[0])
             
for it in trange(1, n_iterations):
                    
    L[it] = X.T @ X +  L[0]
    nu[it] = nu[0] + Y.shape[0]
             
    Mn[it] = inv(L[it]) @ (X.T @ Y + L[0] @ Mn[0])

    V[it] = V[0] + (Y - X @ Mn[it]).T @ (Y - X @ Mn[it]) +\
    (Mn[it] - Mn[0]).T @ L[0] @ (Mn[it] - Mn[0])

    Q[it] = invwishart.rvs(df = nu[it], scale =  V[it])
    A[it] = matrix_normal.rvs(mean = Mn[it], rowcov = inv(L[it]), colcov = Q[it])
```

## Showing results

```python
A_inf = A[burn_out:,:-1,:]     #inferred A matrix for all the iterations after burn_out
b_inf = A[burn_out:,-1,:]      #inferred B matrix for all the iterations after burn_out 
Q_inf = Q[burn_out:]
```

```python
#results for A

n_bins = 6

fig,axs = plt.subplots(2,2,figsize = (15,10))
for i in range(M):
    for j in range(M):
    
        hist, bin_edges = np.histogram(A_inf[:,i,j], density = True, bins = n_bins)
        bin_centers = (bin_edges[1:] + bin_edges[:-1])/2
        labels = np.array([str(round(l,4)) for l in bin_edges])
        w = (max(bin_edges) - min(bin_edges))/n_bins
    
        axs[i,j].set_title('A[%d,%d]' %(i,j))
        axs[i,j].bar(bin_centers, hist, color = 'blue', alpha = 0.3, edgecolor = 'red', linewidth = 2, width = w, align = 'edge')
        axs[i,j].set(xticks=bin_edges, xticklabels=labels)
    
plt.show()
    
```

```python
#results for Q

fig,axs = plt.subplots(2,2,figsize = (15,10))
for i in range(M):
    for j in range(M):
    
        hist, bin_edges = np.histogram(Q_inf[:,i,j], density = True, bins = n_bins)
        bin_centers = (bin_edges[1:] + bin_edges[:-1])/2
        labels = np.array([str(round(l,4)) for l in bin_edges])
        w = (max(bin_edges) - min(bin_edges))/n_bins
    
        axs[i,j].set_title('Q[%d,%d]' %(i,j))
        axs[i,j].bar(bin_centers, hist, color = 'blue', alpha = 0.3, edgecolor = 'red', linewidth = 2, width = w, align = 'edge')
        axs[i,j].set(xticks=bin_edges, xticklabels=labels)
    
plt.show()
```

```python
#results for b

fig,axs = plt.subplots(M,figsize = (10,10))
for i in range(M):
    
    hist, bin_edges = np.histogram(b_inf[:,i], density = True, bins = n_bins)
    bin_centers = (bin_edges[1:] + bin_edges[:-1])/2
    labels = np.array([str(round(l,4)) for l in bin_edges])
    w = (max(bin_edges) - min(bin_edges))/n_bins
    
    axs[i].set_title('b[%d]' %i)
    axs[i].bar(bin_centers, hist, color = 'blue', alpha = 0.3, edgecolor = 'red', linewidth = 2, width = w, align = 'edge')
    axs[i].set(xticks=bin_edges, xticklabels=labels)
    
plt.show()
```
